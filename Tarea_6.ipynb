```
#Tarea 6 Lista Circular
#Ledesma Briones Antonio Yamir

class Nodo:
    def __init__( self , dato, siguiente = None):
        self.dato = dato
        self.siguiente = siguiente
        


class CircularList( ):
#MÉTODO CONSTRUCTOR
    def __init__(self):
        self.__ref = None
        self.__size = 0


#MÉTODO IS EMPTY
    def is_empty(self):
        return self.__size == 0


#MÉTODO INSERT
    def insert(self, value):
        curr_node = self.__ref
        if curr_node == None:
            nuevo = Nodo(value)
            self.__ref = nuevo
            nuevo.siguiente = nuevo
            self.__size += 1

        elif value > curr_node.dato:
            anterior = curr_node.siguiente
            nuevo = Nodo(value)
            curr_node.siguiente = nuevo
            curr_node.siguiente.siguiente = anterior
            self.__ref = curr_node.siguiente
            self.__size += 1
        elif value < curr_node.dato:

            if value < curr_node.siguiente.dato:
                nuevo = Nodo(value)
                siguiente = curr_node.siguiente
                curr_node.siguiente = nuevo
                curr_node.siguiente.siguiente = siguiente
                self.__size += 1

            else:
                nuevo = Nodo(value)
                curr_node = self.__ref.siguiente

                while curr_node != self.__ref:
                    if curr_node.dato < value and curr_node.siguiente.dato > value:
                        siguiente = curr_node.siguiente
                        curr_node.siguiente = nuevo
                        curr_node.siguiente.siguiente = siguiente
                        self.__size += 1
                        break
                    curr_node = curr_node.siguiente
        print(f"Se insertó el valor {value}")
        self.transversal()


#MÉTODO TRANSVERSAL
    def transversal( self ):  
        curr_node = self.__ref

        if curr_node == None:
            print("La lista está vacia")

        else: 
            curr_node = curr_node.siguiente
            while curr_node != self.__ref:
                print(f"{ curr_node.dato } --->", end ="")
                curr_node = curr_node.siguiente
        print(f"{self.__ref.dato} --->")
        print("")


#MÉTODO SEARCH
    def search(self, value ):
        curr_node = self.__ref
        i=0
        if self.__ref != None:
            for i in range(self.__size):
                if curr_node.dato == value:
                    i=i+1
                    return print(f"El valor: {value} Se encuentra en la posición: {i-1}")
                curr_node = curr_node.siguiente
            return print(f"El valor: {value} No se encuentra en la Lista")


#MÉTODO REMOVE
    def remove(self,value):
        curr_node = self.__ref  
        if self.__ref != None:
            for i in range(self.__size):
                if curr_node.siguiente.dato == value:
                    curr_node.siguiente = curr_node.siguiente.siguiente
                    self.__size -= 1
                    return None
                else:
                    curr_node = curr_node.siguiente
                print(f"Se eliminó el valor {value}")      
```
