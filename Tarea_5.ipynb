```
# Tarea 5 Listas Doblemente Ligadas
# Ledesma Briones Antonio Yamir

class DoubleNode:
    def __init__( self , dato, anterior = None, siguiente = None):
        self.dato = dato
        self.siguiente = siguiente
        self.anterior = anterior

class DoubleLinkedList:
#MÉTODO CONSTRUCTOR
    def __init__(self):
        self.__head = None
        self.__tail = None
        self.__size = 0
    

#MÉTODO GET SIZE
    def get_size(self):
        return print(f"El tamaño de la lista es de {self.__size}")
	

#MÉTODO IS EMPTY
    def is_empty(self):
        return self.__size == 0
    

#MÉTODO APPEND
    def append(self,value):
        if self.is_empty():
            nuevo = DoubleNode(value)
            self.__head = nuevo
            self.__tail = nuevo
        else:
            nuevo = DoubleNode(value,self.__tail,None)
            self.__tail.siguiente = nuevo #tail.next = nuevo
            self.__tail = nuevo
        self.__size += 1

        print(f"El valor agregado es: {value}")
        self.transversal()


#MÉTODO FIND FROM TAIL  
    def find_from_tail(self, value ):
        curr_node = self.__tail
        i = 0
        while( curr_node != None ):
            if curr_node.dato == value:
                return print(f"El valor {value} se encuentra en la posición {i+1} desde la cola\n")
            i += 1
            curr_node = curr_node.anterior
        return 


#MÉTODO FIND FROM HEAD
    def find_from_head(self, value ): 
        curr_node = self.__head
        i = 0
        while( curr_node != None ):
            if curr_node.dato == value:
                return print(f"El valor {value} se encuentra en la posición {i+1} desde la cabeza\n")
            i += 1
            curr_node = curr_node.siguiente
            
        return


#MÉTODO REMOVE FROM HEAD
    def remove_from_head(self, value):
        if self.is_empty():
            pass
        else:
            curr_node= self.__head
            while curr_node.dato != value and curr_node != None:
                curr_node = curr_node.siguiente
            if curr_node.dato == value:
                curr_node.anterior.siguiente = curr_node.siguiente
                curr_node.siguiente.anterior = curr_node.anterior
        self.__size -= 1
        print(f"Se removió el {value}")
        self.transversal()

#MÉTODO REMOVE FROM TAIL
    def remove_frome_tail(self , value):
        if self.is_empty():
            pass
        else:
            curr_node = self.__tail
            while curr_node != value and curr_node !=None:
                curr_node = curr_node.anterior
            if curr_node.dato == value:
                curr_node.siguiente.anterior = curr_node.anterior
                curr_node.anterior.siguiente = curr_node.siguiente
        self.size -= 1
        print(f"Se removió el {value}")
        self.transversal()


#MÉTODO TRANSVERSAL 
    def transversal(self):
        curr_node = self.__head
        if curr_node != None:
            while curr_node != None:
                print("<--" ,curr_node.dato, "-->" , end="")
                curr_node = curr_node.siguiente
            print('\n')
        else:
            print('Lista vacia')


#MÉTODO REVERSE TRANSVERSAL
    def reverse_transversal(self):
        curr_node = self.__tail
        if curr_node != None:
            while( curr_node != None ):
                print("<--" ,curr_node.dato, "-->" , end="")
                curr_node = curr_node.anterior
            print('\n')
        else:
            print('Lista vacia')
```

```
ldl = DoubleLinkedList()
ldl.append(90)
ldl.append(80)
ldl.append(70)
ldl.append(60)
ldl.append(50)
ldl.append(40)
ldl.append(30)
ldl.remove_from_head(50)
ldl.find_from_head(70)
ldl.find_from_tail(70)
ldl.get_size()
ldl.reverse_transversal()
```
