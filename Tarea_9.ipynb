```
#ANTONIO YAMIR LEDESMA BRIONES

class PriorityQueue:
    def __init__(self):
        self.__data = [[]]
        self.__size = 0


    def is_empty(self):
        return len(self.__data) == 0


    def length(self):
        return len(self.__data)
  

    def enqueue(self,prioridad , elem):    
        if (type(prioridad) is not int) or (prioridad < 0) :
            print("Valor inválido. Digita un valor válido")
        else:
            if prioridad > len(self.__data):
                for i in range(prioridad - len(self.__data)+3):
                    self.__data.append([])
            self.__data[prioridad].append(elem)
            self.__size += 1

        
    def dequeue(self):
        for i in self.__data:
            if len(i) > 0:
                self.__size -= 1
                if self.__size == 0:
                    print("Ya no hay nadie más en la cola")
                return print(f"{i.pop(0)} fue evacuado")    
        if self.__size == 0:
            print("Nadie más puede evacuar, porque todas las personas han sido evacuadas")


    def to_string(self):
        cont= 0
        for i in self.__data:
            for j in i:
                if cont == 0:
                    print(f"|{j}| <--- head")
                elif cont == self.__size-1:
                    print(f"|{j}| <--- tail")
                else:
                    print(f"|{j}|")
                cont+=1
        print(" ")
```
