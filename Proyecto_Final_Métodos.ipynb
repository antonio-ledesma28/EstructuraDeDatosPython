```
#Método de Gauss-Seidel
import numpy
print ('Método de Gauss-Seidel')
m=int(input('Número de Renglones:')) #Renglones = m
n=int(input('Número de columnas:')) #Columans = n
matrix = numpy.zeros((m,n))
x=numpy.zeros((m))

vector=numpy.zeros((n))
comp=numpy.zeros((m))
error=[]

print ('Introduce los valores')
for r in range(0,m):
    for c in range(0,n):
        matrix[(r),(c)]=float(input("Elemento a["+str(r+1)+str(c+1)+"]: "))
    vector[(r)]=float(input('Vector solución['+str(r+1)+']: '))
tol=float(input("Ingrese la tolerancia deseada: "))
itera=int(input("Número de iteraciones: "))
#Método de Gauss seidel
k=0
while k< itera:
    suma=0
    k=k+1
    for r in range(0,m):
        suma=0
        for c in range(0,n):
            if (c != r):
                suma=suma+matrix[r,c]*x[c]               
        x[r]=(vector[r]-suma)/matrix[r,r]
        print("x["+str(r)+"]: "+str(x[r]))
    del error[:]

    #Comprobación
    for r in range(0,m):
        suma=0
        for c in range(0,n):
            suma=suma+matrix[r,c]*x[c] 
        comp[r]=suma
        dif=abs(comp[r]-vector[r])   
        error.append(dif) 
        print("Error en x[",r,"]= ",error[r])  
    print("Iteraciones: ",k)        
    if all(i<=tol for i in error) == True:
        break
```

```
# Método de Bisección
# Función
def f(x):
    return (x**4)+(3*x**3)-2

# Método de Bisección
def bisection(x0,x1,e):
    step = 1
    print('\n\n*** Método de Bisección ***')
    condition = True
    while condition:
        x2 = (x0 + x1)/2
        if(step==1):    
            print(f'x = {x2} y f(x) = {f(x2)}')
        elif(step==2):    
            print(f'x2 = {x2} y f(x2) = {f(x2)}')
        else:
            print(f'Iteration-%d, x2 = %0.6f and f(x2) = %0.6f' % (step-2, x2, f(x2)))

        if f(x0) * f(x2) < 0:
            x1 = x2
        else:
            x0 = x2
        
        step = step + 1
        condition = abs(f(x2)) > e

    print('\nLa raíz es: %0.8f' % x2)


# Input
x0 = input('Valor inicial: ')
x1 = input('Valor final: ')
e = input('Error Tolerable: ')

#  Input a float
x0 = float(x0)
x1 = float(x1)
e = float(e)

# Condiciones de validación
if f(x0) * f(x1) > 0.0:
    print('Los valores dados no sirven para encontrar la raíz.')
    print('Intenta de nuevo con distintos valores.')
else:
    bisection(x0,x1,e)
```
